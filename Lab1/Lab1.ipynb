```python
'''
Задание 7. Получить решение x заданной системы линейных уравне-
ний вида A* x = B, где A – квадратная матрица n n, B – вектор размерно-
сти n. Решение данной системы уравнений можно получить, набрав строку
кода
x =inv(A)*B,
причем предварительно следует задать значения матрицы А и вектора В,
функция inv() – вычисляет обратную матрицу для матрицы A.
Для случая A=[2 1; 3 4] и B=[4; 11] решите уравнение A* x = B. Обра-
тите внимание, что вектор B – состоит из одного столбца (используется
разделитель столбцов матрицы точка с запятой (;))
'''

A = np.array([[2, 1], [3, 4]])
B = np.array([4, 11]).T

print(np.dot(np.linalg.inv(A), B), '\n')
print(np.linalg.solve(A, B), '\n')
```

```python
'''
Задание 9. Создайте файл test_3Dgraphic.m, в котором наберите при-
веденный выше код п. 12.
'''

def vrosenbrock(x, y):
    return 100 * (y - x ** 2) ** 2 + (1 - x) ** 2


xs = np.arange(-5, 5, 0.05)
ys = np.arange(-5, 5, 0.05)

x, y = np.meshgrid(xs, ys)
z = vrosenbrock(x, y)

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

surf = ax.plot_surface(x, y, z)

fig.colorbar(surf)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

ax.view_init(elev = 90, azim = 90)

plt.show()
```

```python
'''
Задание 10. Создайте новый m-файл и сохраните его под именем
my_func. В файле поместите размещенное выше описание функции
my_func.
'''

X_left, X_right, Y_left, Y_right, N = 0, 5, 6, 10, 1000

x = X_left + np.random.uniform(0, 1, size = N) * (X_right - X_left)
y = Y_left + np.random.uniform(0, 1, size = N) * (Y_right - Y_left)

plt.scatter(x, y)
plt.show()
```

```python
'''
Задание 11. Создайте файл test_my_func.m, в котором наберите код,
демонстрирующий работу функции my_func. Запустите программу и про-
смотрите результат ее работы.
Создайте заголовки гистограмм и подписи по осям с помощью функ-
ций title, xlabel, ylabel, добавив их после функции bar.
'''

BinNumber = 20
k = np.arange(0, BinNumber)

X_bin = X_left + k * (X_right - X_left) / BinNumber
Y_bin = Y_left + k * (Y_right - Y_left) / BinNumber

plt.figure(figsize = (8, 4))

plt.subplot(1, 2, 1)
plt.hist(x, bins = X_bin)
plt.title('X hist')
plt.xlabel('X values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(y, bins = Y_bin)
plt.title('Y hist')
plt.xlabel('Y values')
plt.ylabel('Frequency')

plt.show()
```

```python
'''
Задание 12. Создайте функцию my_gauss_gen, которая генерирует за-
данное количество случайных величин N, имеющих гауссовское распре-
деление с заданным математическим ожиданием m и дисперсией D. По-
стройте гистограмму (рис. 1.3).
'''

m, var = 0, 1

X = np.random.normal(m, var, size = N)

BinNumber = 20
k = np.arange(0, BinNumber)
bin = np.min(X) + k * (np.max(X) - np.min(X)) / BinNumber

plt.hist(X, bins = bin)
plt.title('Normal distribution hist')
plt.xlabel('X values')
plt.ylabel('Frequency')
plt.show()
```

```python
'''
Задание 13. Создайте файл test_cycle.m, в котором сформируйте мат-
рицу Q произвольного размера и заполните ее случайными целыми чис-
лами. Выведите в главном окне MATLAB (Command Window) матрицу Q
(добавьте в код команду disp(Q)).
'''

n, m = 10, 3

Q = np.zeros((n, m))
for i in range(len(Q)):
    for j in range(len(Q[i])):
        Q[i, j] = np.round(np.random.uniform(0, 1) * 10)

print(f'Q is : \n{Q}')
```

```python
'''
Задание 14. В файле test_cycle.m реализуйте цикл, позволяющий вы-
числять сумму элементов матрицы Q. Весь необходимый код напишите в
файле test_cycle.m.
'''

s = 0
for i in Q:
    for j in i:
        s += j

print(f'Sum: {s}, {np.sum(Q)}')
```
